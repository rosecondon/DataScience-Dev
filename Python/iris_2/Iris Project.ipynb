### Supervised Learning: Simple Linear Regression

Base Steps of using Scikit-learn API:

 - Choose a class of model
 - Choose method hyperparaneters
 - Arrange data into features matrix and target array
 - Fit model to data
 - Apply trained model to new data

```python
# x is feature matrix, 10 times the random number btw 0~1
x = 50 * np.random.rand(100)
```

```python
# Design target array (targrt feature), using normaly distributed => randn
y = -5*x+np.random.rand(100)
```

```python
plt.scatter(x,y)
```

```python
# Steps starting
# Choose model: LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
# Choose method hyperparaneters - some consideration
model = LinearRegression(fit_intercept=True)
```

```python
model
```

```python
# Arrange data into features matrix (x) and target array (X)
## feature matrix => ensure it is matrix format. We only have a vector x, use reshape to transfer to matrix
X = x.reshape(-1, 1)
X.shape
```

```python
type(X)
```

```python
# Fit model to data => using fix method, fitting X and y (our feature)
model.fit(X, y)
```

```python
## view the model coefficient to see if it close our initial targrt feature => the number 3 from y=3*x+np.random
model.coef_
```

```python
## the number 2.98668362 is close what we initial "3"
```

```python
## let's see the intercept
model.intercept_
```

```python
## I expect the intercept is close to 0, since it is random, the output is close enough
```

```python
# Apply trained model to new data => predict the performance on unknown data
x_fit = np.linspace(-1, 11) # create unknown data using linspace mothod
```

```python
## target array (X), ensure it is matrix
X_fit = x_fit.reshape(-1, 1)
X_fit.shape
```

```python
## predict model performance on unknown data x_fit. Here we are using our model to predict y_fit with unknown data x_fit
y_fit = model.predict(X_fit)
```

```python
## view our new prediction(x_fit, y_fit) with initial features matrix (x) and feature target (y)
plt.scatter(x,y)
plt.plot(x_fit,y_fit)
```
